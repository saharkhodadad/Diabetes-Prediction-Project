### Diabetes Prediction â€“ Model Inference

This cell loads the trained Logistic Regression pipeline (`diabetes_lr_model.pkl`)
and runs a sample prediction using predefined input values.
The model output indicates whether the given instance corresponds to **Diabetes** or **No Diabetes**.


```python
import joblib
import pandas as pd
import warnings
warnings.filterwarnings('ignore')


loaded_pipe_lr = joblib.load('./model/diabetes_lr_model.pkl')

example = [[2, 101, 90, 21.80, 0.15, 22]]
example_df = pd.DataFrame(example, columns=['Pregnancies', 'Glucose', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age'])

prediction = loaded_pipe_lr.predict(example_df)

print(f"The predicted outcome for the given instance is: {prediction[0]}")
if prediction[0] == 0:
    print("This indicates 'No Diabetes'.")
else:
    print("This indicates 'Diabetes'.")
```
